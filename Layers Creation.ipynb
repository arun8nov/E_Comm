```python
def Reset_Catalog_Schemas(catalog_name:str):
    spark.sql(f"DROP CATALOG IF EXISTS {catalog_name} CASCADE;")
    spark.sql(f"CREATE CATALOG  IF NOT EXISTS {catalog_name};")
    spark.sql(f"USE CATALOG {catalog_name};")
    spark.sql("DROP DATABASE IF EXISTS default CASCADE")
    spark.sql(f"CREATE DATABASE IF NOT EXISTS {bronze};")
    spark.sql(f"CREATE DATABASE IF NOT EXISTS {silver};")
    spark.sql(f"CREATE DATABASE IF NOT EXISTS {gold};")
    spark.sql("CREATE DATABASE IF NOT EXISTS oldp_data")
    spark.sql("USE oldp_data ")
    spark.sql("CREATE VOLUME IF NOT EXISTS raw")
    print("✅ Catalog Reset and Medalian schemas Creation Completed")
    dbutils.fs.cp(
        "/Workspace/Users/arunbabuprakash@gmail.com/E_Comm/raw-data",
        "/Volumes/e_comm/oldp_data/raw/",
        recurse=True
    )
    print("✅ ✅ ✅ Raw Data Uploaded to OLTP Data Volume")
```
